## Generator sztucznych danych na potrzeby pracy z bazą danych o dowolnym zadanym rozmiarze.

```python
import numpy as np
import pandas as pd
from faker import Faker
from IPython.display import display

# tworzę polskojęzyczny egzemplarz/instancję generatora
fake = Faker('pl-PL')
```

#### Listy wybranych ras kotów i psów oraz listy imion zwierzaków umieściłem w repozytorium na GitHubie. Do katalogu ALO pobiorą się 4 pliki.

```python
!git clone https://github.com/jdrapala/ALO.git
```

```python
# wczytuję z plików często spotykane imiona i rasy kotów i psów
imiona_kocie = [imie_kota.strip() for imie_kota in open('ALO//imiona_kotów.txt', 'r', encoding='utf-8')]
imiona_psie = [imie_psa.strip() for imie_psa in open('ALO//imiona_psów.txt', 'r', encoding='utf-8')]
rasy_kocie = [rasa_kota.strip() for rasa_kota in open('ALO//rasy_kotów.txt', 'r', encoding='utf-8')]
rasy_psie = [rasa_psa.strip() for rasa_psa in open('ALO//rasy_psów.txt', 'r', encoding='utf-8')]
```

```python
ile_osob = 100    # tyle osób będzie liczyć plik

# na tej liście będą dokładane kolejne osoby ze swoimi zwierzakami lub bez
rekordy = []

for _ in range(ile_osob):

    plec = np.random.choice(['Kobieta','Mężczyzna'], p=[0.4, 0.6])
    # losuję ile kotów ma osoba
    ile_kotow = np.random.choice([0, 1, 2], p=[0.5, 0.3, 0.2])
    # losuję ile psów ma osoba
    ile_psow = np.random.choice([0, 1, 2], p=[0.5, 0.3, 0.2])

    # losuję osobie płeć i wypełniam zgodnie z nią dane osoby
    if plec == 'Kobieta':
        osoba = {
            "osoba_imię": fake.first_name_female(),
            "osoba_nazwisko": fake.last_name_female(),
            "osoba_wzrost": int(np.random.normal(170,5)),
            "osoba_waga": int(np.random.normal(70,5)),
            "osoba_płeć": plec,
        }
    elif plec == 'Mężczyzna':
        osoba = {
            "osoba_imię": fake.first_name_male(),
            "osoba_nazwisko": fake.last_name_male(),
            "osoba_wzrost": int(np.random.normal(170,5)),
            "osoba_waga": int(np.random.normal(70,5)),
            "osoba_płeć": plec,
        }

    # generuję koty osobie
    for _ in range(ile_kotow):
        kot = {
            'zwierzę_imię': np.random.choice(imiona_kocie),
            'zwierzę_gatunek': 'kot',
            'zwierzę_rasa': np.random.choice(rasy_kocie),
            'zwierzę_wiek': np.random.randint(0,18),
            'zwierzę_czy_żyje': np.random.choice(['tak','nie','null'], p=[0.7,0.2, 0.1]),
        }
        rekordy.append(osoba|kot)
    # generuję psy osobie
    for _ in range(ile_psow):
        pies = {
            'zwierzę_imię': np.random.choice(imiona_psie),
            'zwierzę_gatunek': 'pies',
            'zwierzę_rasa': np.random.choice(rasy_psie),
            'zwierzę_wiek': np.random.randint(0,18),
            'zwierzę_czy_żyje': np.random.choice(['tak','nie','null'], p=[0.7,0.2, 0.1]),
        }
        rekordy.append(osoba|pies)

    # jeżeli osoba nie ma żadnych zwierząt, to trzeba dodać jej jeden rekord z pustymi polami dla zwierzaka
    if ile_kotow + ile_psow == 0:
        rekordy.append(osoba|{
            'zwierzę_imię': 'null',
            'zwierzę_gatunek': 'null',
            'zwierzę_rasa': 'null',
            'zwierzę_wiek': 'null',
            'zwierzę_czy_żyje': 'null',
        })

# pomieszam kolejność, żeby zmusić się do porządniejszego algorytmu indeksowania
np.random.shuffle(rekordy)

# będzie ładniej wyświetlić rekordy jako Ramkę Danych
display(pd.DataFrame(rekordy))

# Zapisuję dane do pliku tekstowego
with open('dane.txt', 'w', encoding='utf-8') as plik:
    for rekord in rekordy:
        # każdy element zamieniam na str, żeby połączyć wszystko i rozdzielić tabulatorami
        plik.write('\t'.join([str(elem) for elem in rekord.values()])+'\n')
```
